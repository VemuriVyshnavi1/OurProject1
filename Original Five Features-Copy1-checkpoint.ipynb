```python
ML_Model = []
accuracy = []
precision = []
recall = []
f1score = []
spe = []
auc = []
tt = []



def storeResults(model, a, b, c, d, e, f, g):
    ML_Model.append(model)
    accuracy.append(round(a, 3))         
    precision.append(round(b, 3))
    recall.append(round(c, 3))
    f1score.append(round(d, 3))
    spe.append(round(e, 3))
    auc.append(round(f, 3))
    tt.append(round(g, 4))
```

# ANOVA

```python
from scipy.stats import f_oneway

group_col = 'Sleep Disorder'

features = ['Gender','Age','Occupation','Sleep Duration','Quality of Sleep','Physical Activity Level','Stress Level','BMI Category'
            ,'Blood Pressure','Heart Rate','Daily Steps']

alpha = 0.05

for feature in features:
    
    subset = data[[group_col, feature]].dropna()
    groups = [group[feature].values for name, group in subset.groupby(group_col)]
    
    f_stat, p_value = f_oneway(*groups) # ANOVA
    
    if p_value < alpha:
        result = "Statistically significant (reject H₀)"
    else:
        result = "Not statistically significant (fail to reject H₀)"
    print(f"{feature}: p-value = {p_value} → {result}")
```
